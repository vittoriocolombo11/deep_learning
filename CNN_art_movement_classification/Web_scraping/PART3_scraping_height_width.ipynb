```python
from web_scraping import Scraper
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd
import numpy as np

#========================================================================
# IMPORTING STUFF
#========================================================================

from web_scraping import Scraper
from bs4 import BeautifulSoup
import requests
import os
import time
import pandas as pd
import numpy as np

#========================================================================
# READING DATA
#========================================================================

BASE_DIR = os.getcwd() + "/temp_txt" # YOUR DIRECTORY HERE
filename = "part4.csv" # YOUR FILE NAME HERE

df2 = pd.read_csv(BASE_DIR+"/"+filename).iloc[:,1:]

#========================================================================
# INSTANTIATING THE SCRAPER OBJECT
#========================================================================

driver_path = "YOUR PATH OF CHROME DRIVER HERE" # YOUR PATH OF CHROME DRIVER HERE
scraper = Scraper(driver_path) # here you'll see the driver open

CHROMIUM_CHANGE_RATE = 100 # put this to 100

OUT_FILE_NAME = "part4_updated.csv" # YOUR FILE NAME HERE
list_of_jpeg = []
list_of_height = []
list_of_width = []

#========================================================================
# SCRAPING WIDTH, HEIGHT AND JPEG LINK
#========================================================================

for i in range(len(df2)):
    
    # every 100 pictures we close the driver and open another one
    if i % CHROMIUM_CHANGE_RATE==0 and i!=0:
        scraper.driver.close() # close the open driver
        scraper = Scraper(driver_path) # opens another one
        print(scraper.driver.execute_script("return navigator.userAgent;")) # prints for your info the user agent
    try:
        url = df2.image_link[i] # the url is the one indexed in the dataframe
        scraper.get_url(url) # connects to the url
        all_images = scraper.driver.find_elements_by_tag_name('img') # these are a bunch of selenium elements: all the
        # images in the page
        # this below is the right element of the picture we're interested in 
        selenium_right_elem = [all_images[i] for i in range(len(all_images)) if 'https://www.copia-di-arte.com/kunst/' in all_images[i].get_attribute("src")][0]
        try:   
            image = selenium_right_elem.get_attribute('src') # link to jpg image
            height = selenium_right_elem.get_attribute('height') # height of the image
            width = selenium_right_elem.get_attribute('width')

            list_of_jpeg.append(image)
            list_of_height.append(height)
            list_of_width.append(width)
        except:
            list_of_jpeg.append(np.nan)
            list_of_height.append(np.nan)
            list_of_width.append(np.nan)
            pass
        
    except:    
        list_of_jpeg.append(np.nan)
        list_of_height.append(np.nan)
        list_of_width.append(np.nan)
        pass
    
#========================================================================
# UPDATING DATAFRAME AND SAVING
#========================================================================
    
df2['jpeg'] = list_of_jpeg
df2['height'] = list_of_height
df2['width'] = list_of_width

df2.to_csv(OUT_FILE_NAME)
```
